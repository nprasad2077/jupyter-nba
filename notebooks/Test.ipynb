```python
import matplotlib.pyplot as plt
import numpy as np
```

# Hello. This is a plot test.

```python
# Create a list of labels for the x-axis (e.g., 'A', 'B', 'C', ..., 'J')
x = list('ABCDEFGHIJ')

# Create a list of random values for the y-axis
y = np.random.randint(10, 50, size=10)
```

## Now, we can create a bar graph using plt.bar()

```python
plt.bar(x, y)

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Bar Graph Title')

# Display the graph
plt.show()
```
